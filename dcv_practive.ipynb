Задание №5

```python
white = 5
black = 7
n = white + black
X = [-1, 0, 1] #случайная величина: -1(2 белых), 0(1 белый и 1 черный), 1(2 черных)
total_end = math.comb(n,2) #общее количество исходов при вытаскивании 2 шаров
p_ww = math.comb(white,2)/total_end #вероятность вытащить 2 белых
p_bb = math.comb(black,2)/total_end #вероятность вытащить 2 черных
wb = 1-(p_ww+p_bb) #вероятность вытащить 1 белого и 1 черного

print(total_end, p_ww, p_bb, wb)
#математическое ожидание
E_X = X[0]*p_ww + X[1]*wb + X[2]*p_bb
print("E_X =", E_X)

xs = X
Fs = [p_ww, p_bb, wb]
xs_plot = [-2] + xs
Fs_plot = [0.0] + Fs



plt.step(xs_plot, Fs_plot, where='post')
plt.ylim(-0.05, 1.05)
plt.xlim(-2, 2)
plt.xlabel('x')
plt.ylabel('F(x) = P(X ≤ x)')
plt.title('Функция распределения случайной величины X')
plt.grid()
#plt.show()

fig, ax = plt.subplots()
ax.step(xs_plot, Fs_plot, where='post')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.scatter(xs, Fs, color='red', s=40, zorder=5)

ax.grid()
```

Суть задания - $P(X=k) = \frac{C}{(k*(k+1))}$, надо найти С, а так же $P(X<=3)$

```python
from sympy import symbols, summation

k = symbols('k', integer=True, positive=True) 
expr = 1/(k*(k+1))
p = summation(expr, (k, 1, 3))
print(p)
print(float(p))
```
