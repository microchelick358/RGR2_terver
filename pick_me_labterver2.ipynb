```python
import numpy as np
import matplotlib.pyplot as plt

# Настройка классического стиля с пастельной палитрой
plt.style.use('default')
colors = ['#FFC1CC', '#9EC1CF', '#9EE09E'] # Розовый, голубой, мятный
true_val_color = '#777777'

k = 2
lambd_true = (k + 1)
n_max = 1000

# Генерация данных
samples = [
    np.random.exponential(scale=lambd_true, size=500),
    np.random.exponential(scale=lambd_true, size=1000),
    np.random.exponential(scale=lambd_true, size=2000)
]

titles = ['n=500', 'n=1000', 'n=2000']
fig, axes = plt.subplots(3, 1, figsize=(10, 12))

for i, ax in enumerate(axes):
    n = len(samples[i])
    x_axis = np.arange(1, n + 1)
    estimate = np.cumsum(samples[i]) / x_axis
    
    ax.plot(x_axis, estimate, color=colors[i], linewidth=2, label='Оценка $\lambda$')
    ax.axhline(y=lambd_true, color=true_val_color, linestyle='--', linewidth=1.5, alpha=0.6, label='Истинное значение')
    
    # Эстетика осей
    ax.set_title(titles[i], fontsize=12, fontweight='bold', color='#444444', loc='left')
    ax.set_xlabel('Объем выборки (n)', fontsize=10, color='#666666')
    ax.set_ylabel('Значение', fontsize=10, color='#666666')
    ax.grid(True, linestyle=':', alpha=0.6)
    ax.legend(frameon=True, facecolor='white', edgecolor='#eeeeee')
    
    # Убираем лишние рамки для легкости
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
